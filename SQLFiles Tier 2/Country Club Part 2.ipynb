```python
from sqlalchemy import create_engine
import pandas as pd
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

```python
Q10 = pd.read_sql_query(
    """SELECT sub.name, SUM(cost) AS revenue
        FROM (
            SELECT f.name, CASE WHEN b.memid > 0 THEN f.membercost 
                                ELSE f.guestcost END AS cost
            FROM `Facilities` AS f 
            INNER JOIN `Bookings` AS b 
                ON f.facid = b.facid 
            INNER JOIN `Members` AS m
                ON b.memid = m.memid) AS sub
        GROUP BY sub.name
        HAVING revenue < 1000
        ORDER BY revenue;""", engine)
Q10
```

```python
Q11 = pd.read_sql_query(
    """SELECT membername, 
        CASE WHEN sub.recommendedby = 0 THEN "None"
            ELSE recommendername END AS recname
        FROM (
            SELECT m.recommendedby, m.surname || ', ' || m.firstname AS membername, 
                r.surname || ', ' || r.firstname AS recommendername
            FROM Members AS m
            LEFT JOIN Members AS r
                ON m.recommendedby = r.memid
            WHERE m.memid > 0
            ORDER BY membername) AS sub""", engine)
Q11
```

```python
Q12 = pd.read_sql_query(
    """SELECT sub.name AS facility, mem.surname || ', ' || mem.firstname AS membername, sub.usage_hours
        FROM (
            SELECT f.name, b.memid, SUM(b.slots * 0.5) AS usage_hours
            FROM `Facilities` AS f
            INNER JOIN `Bookings` AS b
                ON f.facid = b.facid
            INNER JOIN `Members` AS m
                ON b.memid = m.memid
            WHERE m.memid > 0
            GROUP BY f.name, b.memid) AS sub
        INNER JOIN `Members` AS mem
            ON sub.memid = mem.memid
        ORDER BY facility, membername""", engine)
Q12
```

```python
Q13 = pd.read_sql_query(
    """SELECT strftime('%m', b.starttime) AS month, f.name AS facility, SUM(b.slots * 0.5) AS usage_hours
        FROM `Facilities` AS f
        INNER JOIN `Bookings` AS b
            ON f.facid = b.facid
        INNER JOIN `Members` AS m
            ON b.memid = m.memid
        WHERE m.memid > 0
        GROUP BY month, facility""", engine)
Q13
```
